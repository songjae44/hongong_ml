```python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
                # 도미 데이터 생성
```

```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
# 빙어 데이터 생성
```

```python
# 도미, 빙어 데이터 시각화
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 도미, 빙어 데이터를 길이와 무게로 합치기
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
# 물고기의 데이터를 2차원 리스트 형태로 변경
fish_data = [[l, w] for l, w in zip(length, weight)]
```

```python
print(fish_data)
```

```python
# 도미는 숫자 1, 빙어는 숫자 0으로 표현
# 앞의 35개의 데이터는 도미의 데이터, 뒤의 14개 데이터는 빙어의 데이터
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier # k-최근접 이웃 알고리즘 구현을 위한 라이브러리 호출
```

```python
kn = KNeighborsClassifier() # 알고리즘 객체 생성
```

```python
kn.fit(fish_data, fish_target) # fit메서드로 주어진 데이터를 가지고 알고리즘을 훈련
```

```python
# score메서드로 모델을 평가
# 0 ~ 1 사이로 평가지표 출력
kn.score(fish_data, fish_target)
```

k-최근접 이웃 알고리즘
- 어떤 데이터에 대한 답을 구할 때 주위의 다른 데이터를 보고 다수를 차지하는 것을 정답으로 사용
- 근묵자흑과 같이 주위의 데이터로 현재 데이터를 판단
- 새로운 데이터에 대해 예측할 때 가장 가까운 직선거리에 어떤 데이터가 있는지를 살피면 된다
- 단점은 데이터가 아주 많은 경우 사용하기 어렵다(메모리가 많이 필요, 직선 거리 계산에도 많은 시간)


```python
# 알고리즘 메커니즘 확인
kn.predict([[30, 600]]) # 앞의 fit메서드와 같이 2차원 리스트를 전달해야함
# 결과값을 보면 도미의 1을 반환
```

```python
# _fit_X 속성에 fish_data 모두 가지고 있음
print(kn._fit_X)
```

```python
# _y 속성에는 fish_target을 가지고있음
print(kn._y)
```

```python
# n_neighbors 매개변수로 몇 개의 데이터를 참고할 지 정할 수 있음(기본 5)
kn49 = KNeighborsClassifier(n_neighbors=49) # 참고 데이터를 49개로 한 kn49 모델
```

```python
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
# fish_data의 데이터 49개 중 도미가 35개로 다수를 차지하므로 어떤 데이터를 넣어도 도미로 예측
kn49.predict([[1, 45]])
```

```python
# 35/49의 결과가 정확도와 일치
# 도미만 올바르게 맞히기 때문
print(35/49)
```
