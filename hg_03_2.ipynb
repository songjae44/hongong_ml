```python
import numpy as np

perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

```python
# 데이터를 훈현 세트와 테스트 세트로 분리
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state=42)

# 특성 데이터는 2차원 배열로 변환
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)
```

```python
# 최근점 이웃 개수를 3으로 하는 모델을 훈련
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor(n_neighbors=3)
knr.fit(train_input, train_target)
```

```python
# 모델을 사용해 길이가 50cm인 농어의 무게를 예측
print(knr.predict([[50]]))
```

```python
# 훈련 세트와 50cm 농어 그리고 이 농어의 최근접 이웃을 산점도에 표시
import matplotlib.pyplot as plt

# 50cm 농어의 이웃을 구한다
distances, indexes = knr.kneighbors([[50]])

# 훈련 세트의 산점도를 그린다
plt.scatter(train_input, train_target)

# 훈련 세트 중에서 이웃 샘플만 다시 그린다
plt.scatter(train_input[indexes], train_target[indexes], marker='D')

# 50cm 농어 데이터
plt.scatter(50, 1033, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 이웃 샘플의 평균
print(np.mean(train_target[indexes]))
```

```python
# 길이가 100인 농어 무게 예측
print(knr.predict([[100]]))
```

```python
# 100cm 농어의 이웃을 구한다
distances, indexes = knr.kneighbors([[100]])

# 훈련 세트의 산점도를 그린다
plt.scatter(train_input, train_target)

# 훈련 세트 중에서 이웃 샘플만 다시 그린다
plt.scatter(train_input[indexes], train_target[indexes], marker='D')

# 100cm 농어 데이터
plt.scatter(100, 1033, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

### 선형회귀
***선형회귀***는 널리 사용되는 대표적인 회귀 알고리즘
- 비교적 간단하고 성능이 뛰어남
- 특성이 하나인 경우 어떤 직선을 학습하는 알고리즘

```python
# 사이킷런의 선형회귀 모델 호출
from sklearn.linear_model import LinearRegression

# 모델을 훈련
lr = LinearRegression()
lr.fit(train_input, train_target)

# 50cm 농어에 대해 예측
print(lr.predict([[50]]))
```

선형회귀 직선은 $y = a * x + b$처럼 쓸 수 있다

데이터에 가장 잘 맞는 a와 b를 모델 객체의 coef_와 intercept_속성에 저장

```python
# 기울기와 y절편 값 출력
print(lr.coef_, lr.intercept_)
```

```python
# 농어의 길이 15에서 50까지 직선으로 그리기

# 훈련세트의 산점도 그리기
plt.scatter(train_input, train_target)

# 15cm에서 50cm까지 1차 방정식 그래프를 그리기
plt.plot([15, 50], [15*lr.coef_+lr.intercept_, 50*lr.coef_+lr.intercept_])

# 50cm 농어 데이터
plt.scatter(50, 1241.8, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 훈련 세트와 테스트 세트에 대한 결정계수 점수 확인
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

```python
# 2차 방정식(곡선) 형태로 만들기 위해 길의의 제곱 컬럼 추가
# column_stack()함수를 사용
train_poly = np.column_stack((train_input ** 2, train_input))
test_poly = np.column_stack((test_input ** 2, test_input))
# train_input**2 식에도 numpy 브로드캐스팅 적용됨
```

```python
# 새롭게 만든 데이터셋의 크기 확인
print(train_poly.shape, test_poly.shape)
```

```python
# train_poly를 사용한 선형회귀 모델 학습
# 이때 타깃값은 그대로 사용
lr = LinearRegression()
lr.fit(train_poly, train_target)

print(lr.predict([[50**2, 50]]))
```

```python
# 모델이 훈련한 계수와 절편을 출력
print(lr.coef_, lr.intercept_)
```

모델이 학습한 그래프

$무게 = 1.01 * 길이^2 - 21.6 * 길이 + 116.05$

```python
# 구간별 직선을 그리기 위해 15에서 49까지 정수 배열 만들기
point = np.arange(15, 50)

# 훈련 세트의 산점도를 그리기
plt.scatter(train_input, train_target)

# 15~49까지 2차 방정식 그래프 그리기
plt.plot(point, 1.01*point**2 - 21.6*point + 116.05)

# 50cm 농어 데이터
plt.scatter([50], [1574], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 훈련 세트와 테스트 세트의 결정계수 점수 출력
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))
```

k-최근접 이웃 회귀를 사용해서 예측할 때 문제점
- 훈련 세트 범위 밖의 샘플을 예측할 수 없다는 점
- k-최근점 이웃 회귀는 아무리 멀리 떨어져 있어도 가장 근접한 이웃 샘플의 타깃을 평균하여 예측
